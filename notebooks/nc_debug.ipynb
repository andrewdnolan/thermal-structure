```python
import numpy as np 
import xarray as xr 
import matplotlib.pyplot as plt

with xr.open_dataset('../results/test.nc') as src: 
    print(src)
```

```python
src["Z"] = src["zbed"]  + (src.Y * src["height"])
```

```python
fig, ax = plt.subplots(figsize=(30,10))
plt.pcolormesh(src.X, src.Z, np.sqrt(src["velocity 1"]**2 + src["velocity 1"]**2), shading='gouraud')
plt.plot(src.X.values.flatten(), 
         src.Z.values.flatten(),
          'o', color='m', markersize=1)
plt.colorbar()
```
