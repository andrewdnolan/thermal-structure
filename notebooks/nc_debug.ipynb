```python
import numpy as np 
import xarray as xr 
import matplotlib.pyplot as plt

with xr.open_dataset('../results/test.nc') as src: 
    print(src)
```

```python
src["Z"] = src["zbed"]  + (src.Z * src["height"])
```

```python
fig, ax = plt.subplots(figsize=(30,10))
plt.pcolormesh(src.X, src.Z.isel(t=0), np.sqrt(src["velocity 1"].isel(t=0)**2 + src["velocity 1"].isel(t=0)**2), shading='gouraud')
plt.colorbar()
```
